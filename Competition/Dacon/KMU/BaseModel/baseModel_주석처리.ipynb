# 함수 정의

```python
# TOP 5 예측의 RECALL을 계산하는 함수


def recall5(answer_df, submission_df):

    primary_col = answer_df.columns[0] # answer_df의 0번째 열  : 이력서 번호
    secondary_col = answer_df.columns[1] # answer_df의 1번째 열 : 공고번호
                                        ## 열의 이름을 정의했다고 생각

    prediction_counts = submission_df.groupby(primary_col).size()
    # submission 에서 primary_col값이 정확히 5개의 secondary_col 예측을 가져야 하는 것을 확인
    ## 그래서 submission_df에서 primary_col열 기준으로 groupby하고 그룹내의 예측 수 확인
    ### size() 함수는 데이터프레임의 행수 반환

    if not all(prediction_counts == 5): # 5개 확인
        raise ValueError(f"Each {primary_col} should have exactly 5 {secondary_col} predictions.")


    # submission의 예측된 값들에 null값이 있는지 확인
    if submission_df[secondary_col].isnull().any():
        raise ValueError(f"Predicted {secondary_col} contains NULL values.")

    # 예측값에 중복이 있는지 확인
    duplicated_preds = submission_df.groupby(primary_col).apply(lambda x: x[secondary_col].duplicated().any())
    if duplicated_preds.any():
        raise ValueError(f"Predicted {secondary_col} contains duplicates for some {primary_col}.")


    # Filter the submission dataframe based on the primary_col present in the answer dataframe
    # primary_col 즉 resume_seq가 양측에 있는지 확인 후 남김
    # submission_df 와 answer_df 에서 primary_col 값이 일치하는 부분은 남긴다.
    submission_df = submission_df[submission_df[primary_col].isin(answer_df[primary_col])]


    # primary_col 값에 대해 상위 5개 가져옴
    top_5_preds = submission_df.groupby(primary_col).apply(lambda x: x[secondary_col].head(5).tolist()).to_dict()

    # answer_df를 primary_col기준으로 groupby하고 딕셔너리로 변환
    ## 이유는 모르겠음 그냥 리스트로 보기 편하게 하기위해 ?
    ## 반복문 돌릴려고 ?
    true_dict = answer_df.groupby(primary_col).apply(lambda x: x[secondary_col].tolist()).to_dict()


    individual_recalls = [] # primary_col 값에 대한 recall값 저장 리스트
    for key, val in true_dict.items(): #
        if key in top_5_preds:
            correct_matches = len(set(true_dict[key]) & set(top_5_preds[key]))
            # 실제값과 예측값중 얼마나 많은 일치가 있는지
            ## true_dict[key] : 현재 primary_col 값
            ## top_5_preds[key] : primary_col 값에 5개 예측값
            ### set(true_dict[key]) : 중복 제거
            ###  set(top_5_preds[key]) : 중복 제거
            #### set(true_dict[key]) & set(top_5_preds[key]) : 두 값에 대한 교집합
            ##### len(set(true_dict[key]) & set(top_5_preds[key])) : 교집합의 크기 , 즉 일치하는 예측값 수
            individual_recall = correct_matches / min(len(val), 5) # 공정한 평가를 가능하게 위하여 분모(k)를 'min(len(val), 5)' 로 설정함
            individual_recalls.append(individual_recall)
            # 리스트에 추가


    recall = np.mean(individual_recalls) # 평균 recall값 계산 최종반환
    return recall

# ★★★★★ 정리 ★★★★★
# true_dict 변수는 answer_df 데이터프레임을 딕셔너리로 변환 한 것
# 딕셔너리안에는 primary_col 값인 실제값이 들어있음
# top_5_preds 변수는 submission_df 데이터프레임을 primary_col 열 기준으로 groupby한것
# 그리고 각 그룹에서 5개의 예측된 secondary_col 을 딕셔너리로 저장
```

```python
recommendations = [] # 추천결과를 저장할 리스트

for idx, user in enumerate(train_train2.index): # 인덱스와 사용자 반환
    # 해당 사용자가 지원한 채용 공고를 저장
    ## train_train_2에서 사용자가 지원한 공고를 찾고 저장
    applied_jobs = set(train_train2.loc[user][train_train2.loc[user] == 1].index)

    # 해당 사용자의 추천 점수 (높은 점수부터 정렬)
    sorted_job_indices = total_train_user_c[idx].argsort()[::-1]
    # 지원하지 않은 공고중에서 가장 높은 추첝머수를 가진 5개를 recommended_jobs에 저장
    ## 이미 지원한 공고는 제외 : if job not in applied_jobs][:5]
    recommended_jobs = [job for job in train_train2.columns[sorted_job_indices] if job not in applied_jobs][:5]

    for job in recommended_jobs: # 5개 채용공고 반복
        recommendations.append([user, job]) # 사용자와 추천된 공고 리스트에 추가
```
