# Import

```python
import pandas as pd
# ì €í¬ëŠ” Tabular Dataë¥¼ ë‹¤ë£¨ê¸° ë•Œë¬¸ì— ì•„ë˜ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ í˜¸ì¶œí•©ë‹ˆë‹¤.
from autogluon.tabular import TabularDataset, TabularPredictor
```

# Data Load

```python
train_df = pd.read_csv('data/train.csv')
test_df = pd.read_csv('data/test.csv')
```

```python
train_df.head()
```

# Train

autogluon ë‚´ë¶€ì—ëŠ” ì—¬ëŸ¬ê°€ì§€ ëª¨ë¸ì„ ì•Œì•„ì„œ ë¹„êµí•´ì£¼ê³  ì½”ë“œë¡œ êµ¬í˜„í•˜ê¸° ì‰½ìŠµë‹ˆë‹¤.

```python
# autogluon í•™ìŠµì„ ìœ„í•œ ë°ì´í„° í˜•íƒœë¡œ ë³€í™˜
train = TabularDataset(train_df.drop(['ID'], axis=1))
test = TabularDataset(test_df.drop(['ID'], axis=1))
```

```python
# ì´ë ‡ê²Œ í•œ ì¤„ë§Œ ì‘ì„±í•˜ë©´ ë‚´ë¶€ì—ì„œ ì•Œì•„ì„œ í•™ìŠµí•´ì¤ë‹ˆë‹¤.
predictor = TabularPredictor(label='ì „í™”í•´ì§€ì—¬ë¶€', eval_metric='f1_macro',).fit(train)
```

```python
# ê°ê°ì˜ ëª¨ë¸ì˜ í›ˆë ¨ ì„±ëŠ¥ì„ í‰ê°€í•  ìˆ˜ ìˆìŒ
ld_board = predictor.leaderboard(train, silent=True)

ld_board
```

```python
# ì˜ˆì¸¡í•˜ê¸°
pred_y = predictor.predict(test)
```

```python
# ì œì¶œ íŒŒì¼ ìƒì„±
submit = pd.DataFrame()

submit['ID'] = test['ID']
submit['ì „í™”í•´ì§€ì—¬ë¶€'] = pred_y

submit.to_csv('submit.csv', index=False)
```
